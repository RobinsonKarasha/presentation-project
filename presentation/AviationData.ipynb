```python
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame containing the relevant columns

# List of injury types
injury_types = ['Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured']

# Iterate through each injury type and create individual charts
for injury_type in injury_types:
    plt.figure(figsize=(10, 6))
    for category in data['Aircraft.Category'].unique():
        # Filter data for the current category and injury type
        category_data = data[data['Aircraft.Category'] == category]
        plt.plot(category_data['Aircraft.Category'], category_data[injury_type], label=category)

    plt.title(f'Line Graph of {injury_type} by Aircraft Category')
    plt.xlabel('Aircraft Category')
    plt.ylabel(injury_type)
    plt.legend(title='Aircraft Category', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
